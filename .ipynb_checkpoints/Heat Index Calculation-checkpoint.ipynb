```python
def calculate_frequency_and_std(dataframe, group_column):
    # Ensure the "Date/Time" column is in datetime format
    dataframe['Date/Time'] = pd.to_datetime(dataframe.index)
    
    # Calculate the difference in time between consecutive measurements within each group
    dataframe['Time_Diff'] = dataframe.groupby(group_column)['Date/Time'].diff().dt.total_seconds() / 60  # Convert to minutes
    
    # Define a function to get the smallest mode
    def smallest_mode(series):
        return series.mode().min()
    
    # Calculate the average frequency, standard deviation, count, and smallest mode for each group
    results = dataframe.groupby(group_column)['Time_Diff'].agg(['mean', 'std', 'count', ('mode', smallest_mode)]).reset_index()
    
    # Calculate the overall average of means and modal values
    overall_mean = results['mean'].mean()
    overall_std = results['std'].mean()
    overall_mode_avg = results['mode'].mean()
    
    # Print the results for each group
    for index, row in results.iterrows():
        print(f"Site ID: {row[group_column]}, Average Frequency: {row['mean']:.2f} minutes, Standard Deviation: {row['std']:.2f} minutes, Modal Frequency: {row['mode']:.2f} minutes, Data Points: {row['count']}")
    
    # Print the overall averages
    print(f"\nOverall Average Frequency (Mean): {overall_mean:.2f} minutes, Overall Average Standard Deviation: {overall_std:.2f} minutes, Overall Modal Average Frequency: {overall_mode_avg:.2f} minutes")

calculate_frequency_and_std(outdoor_temp, 'site_id')
print("")
# calculate_frequency_and_std(indoor_temp, 'sensor_id_indoor_temp')
# print("")
calculate_frequency_and_std(outdoor_hum, 'site_id')
# print("")
# calculate_frequency_and_std(indoor_hum, 'sensor_id_indoor_hum')
# print("")
```

```python
import pandas as pd

# Assuming outdoor_temp, outdoor_hum, and outdoor_link_df are already defined DataFrames

# Rename 'Value' column to specific names before setting the index
outdoor_temp.rename(columns={'Value': 'Value_temp'}, inplace=True)
outdoor_hum.rename(columns={'Value': 'Value_hum'}, inplace=True)

# Map 'site_id' to 'point_id_outdoor_temp' in outdoor_temp
outdoor_temp['site_id'] = outdoor_temp['point_id_outdoor_temp'].map(outdoor_link_df.set_index('point_id_outdoor_temp')['site_id'])

# Map 'site_id' to 'point_id_outdoor_hum' in outdoor_hum
outdoor_hum['site_id'] = outdoor_hum['point_id_outdoor_hum'].map(outdoor_link_df.set_index('point_id_outdoor_hum')['site_id'])

# outdoor_temp = outdoor_temp.sort_index()
# outdoor_hum = outdoor_hum.sort_index()

# No need to sort by 'Date/Time' index as it should already be in datetime format and sorted by default
# Define the tolerance for merging (e.g., 10 minutes)
tolerance = pd.Timedelta(minutes=10)

# Prepare for merging by resetting the index to bring 'Date/Time' back as a column, facilitating the merge_asof operation
temp_df = outdoor_temp.reset_index()
hum_df = outdoor_hum.reset_index()

# Initialize an empty list to hold the result frames
result_frames = []

for site_id in outdoor_link_df['site_id'].unique():
    temp_site_df = temp_df[temp_df['site_id'] == site_id]
    hum_site_df = hum_df[hum_df['site_id'] == site_id]
    
    # Perform the asof merge for the current site_id
    merged_df = pd.merge_asof(temp_site_df.sort_values('Date/Time'), hum_site_df.sort_values('Date/Time'), 
                              on='Date/Time', by='site_id', tolerance=tolerance, direction='nearest')
    result_frames.append(merged_df)

# Concatenate all the merged data frames for each site_id
final_merged_df = pd.concat(result_frames, ignore_index=True)

# Assuming you want to keep only rows where both temperature and humidity data are available
final_merged_df.dropna(subset=['Value_temp', 'Value_hum'], inplace=True)

# Print the first few rows of the final merged DataFrame
print(final_merged_df[['Date/Time', 'site_id', 'Value_temp', 'Value_hum']].head())
```

```python
tqdm.pandas()

# Convert the humidity values to decimal form
# final_merged_df['Value_hum'] = final_merged_df['Value_hum'] / 100

def calculate_heat_index(row):
    temp_k = row['Value_temp'] + 273.16  # Convert temperature to Kelvin
    RH = row['Value_hum']  # Get the relative humidity

    # Check if both inputs are floats
    if isinstance(temp_k, float) and isinstance(RH, float):
        try:
            heatindex_k = heatindex(temp_k, RH)  # Calculate the heat index
            if isinstance(heatindex_k, float):
                return heatindex_k - 273.16  # Convert heat index back to Celsius
            else:
                return None
        except SystemExit:  # Catch the exception here
            return np.nan
    else:
        return None

# Apply the function to each row in the dataframe
final_merged_df['Outdoor_Heat_Index'] = final_merged_df.progress_apply(calculate_heat_index, axis=1)
```

```python
import pandas as pd

# Assuming indoor_temp, indoor_hum, and indoor_link_df are already defined DataFrames

# Rename 'Value' column to specific names before setting the index
indoor_temp.rename(columns={'Value': 'Value_temp'}, inplace=True)
indoor_hum.rename(columns={'Value': 'Value_hum'}, inplace=True)

# Define the tolerance for merging (e.g., 10 minutes)
tolerance = pd.Timedelta(minutes=10)

# Prepare for merging by resetting the index to bring 'Date/Time' back as a column, facilitating the merge_asof operation
temp_df = indoor_temp.reset_index()
hum_df = indoor_hum.reset_index()

# Initialize an empty list to hold the result frames
result_frames = []

for _, row in indoor_link_df.iterrows():
    point_id_indoor_hum = row['point_id_indoor_hum']
    point_id_indoor_temp = row['point_id_indoor_temp']

    temp_point_df = temp_df[temp_df['point_id_indoor_temp'] == point_id_indoor_temp]
    hum_point_df = hum_df[hum_df['point_id_indoor_hum'] == point_id_indoor_hum]

    # Perform the asof merge for the current point_id_indoor_hum and point_id_indoor_temp
    merged_df = pd.merge_asof(temp_point_df.sort_values('Date/Time'), hum_point_df.sort_values('Date/Time'),
                              on='Date/Time', tolerance=tolerance, direction='nearest')

    # Add the site_id to the merged DataFrame
    merged_df['site_id'] = row['site_id']

    result_frames.append(merged_df)

# Concatenate all the merged data frames for each point_id_indoor_hum and point_id_indoor_temp
final_merged_df = pd.concat(result_frames, ignore_index=True)

# Assuming you want to keep only rows where both temperature and humidity data are available
final_merged_df.dropna(subset=['Value_temp', 'Value_hum'], inplace=True)

# Print the first few rows of the final merged DataFrame
print(final_merged_df[['Date/Time', 'site_id', 'point_id_indoor_temp', 'point_id_indoor_hum', 'Value_temp', 'Value_hum']].head())
```

```python
# Assuming your dataframe has a datetime column named "timestamp"
merged_df['Date'] = merged_df['Date/Time'].dt.date

# Step 1: Day-Level Aggregation with 'site_type_outdoor'
day_level = merged_df.groupby(['sensor_id_indoor_hum', 'Date', 'Day/night']).agg(
    daily_mean_indoor=('Indoor_Value', 'mean'),
    daily_mean_outdoor=('Value', 'mean'),
    daily_max_indoor=('Indoor_Value', 'max'),
    daily_max_outdoor=('Value', 'max'),
    site_type_outdoor=('site_type_outdoor', 'first'),  # Assuming each sensor_id, Date, Day/night combo has a single site type
    mean_ndvi=('mean_ndvi', 'first'),  # Assuming each sensor_id, Date, Day/night combo has a single site type
    mean_ndbi=('mean_ndbi', 'first')  # Assuming each sensor_id, Date, Day/night combo has a single site type
).reset_index()

# Step 2: Sensor-Level Aggregation
aggregated_data = day_level.groupby(['sensor_id_indoor_hum', 'Day/night']).agg(
    average_daily_mean_indoor=('daily_mean_indoor', 'mean'),
    average_daily_mean_outdoor=('daily_mean_outdoor', 'mean'),
    average_daily_max_indoor=('daily_max_indoor', 'mean'),
    average_daily_max_outdoor=('daily_max_outdoor', 'mean'),
    mean_ndvi=('mean_ndvi', 'first'),
    mean_ndbi=('mean_ndbi', 'first')
).reset_index()
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

def annotate_fit(x, y, ax):
    # Drop NaN values
    df = pd.DataFrame({'x': x, 'y': y}).dropna()
    X = sm.add_constant(df['x'])
    model = sm.OLS(df['y'], X).fit()
    a, b = model.params['const'], model.params['x']
    r2 = model.rsquared
    
    # Calculate the confidence interval for R using Pingouin
    ci = pg.corr(x=df['x'], y=df['y']).loc['pearson', 'CI95%']
    # Square the CI values to get the CI for R^2
    ci_r2 = [value**2 for value in ci]
    
    # ax.annotate(f'Y = {a:.2f} + {b:.2f}X\n$R^2$ = {r2:.2f}, 95% CI: [{ci_r2[0]:.2f}, {ci_r2[1]:.2f}]', 
    #             xy=(0.05, 0.85), xycoords='axes fraction', fontsize=14)
    ax.annotate(f'$R^2$ = {r2:.2f}\n95% CI:[{ci_r2[0]:.2f}, {ci_r2[1]:.2f}]', 
                xy=(0.025, 0.875), xycoords='axes fraction', fontsize=14)

# Choose either 'day_level' or 'aggregated_data' for analysis
analysis_level = 'day_level'  # Set to either 'day' or 'aggregated'

if analysis_level == 'day_level':
    analysis_frame = day_level
    axis_prefix = 'Daily'
elif analysis_level == 'aggregated':
    analysis_frame = aggregated_data
    axis_prefix = 'Average Daily'

# day_night = 'Day'  # Change as needed
day_night = 'Night'  # Change as needed

# Sort the analysis_frame by 'mean_ndvi'
analysis_frame = analysis_frame.sort_values('mean_ndvi')

plt.figure(figsize=(10, 10))  # Adjusted for a square plot

# Helper function to find the common limits
def find_common_limits(series1, series2, upper_bound=None):
    combined = pd.concat([series1, series2], ignore_index=True)
    lower = combined.min()
    upper = combined.max()
    if upper_bound is not None:
        upper = min(upper, upper_bound)
    return lower, upper
    
# Find the common limits for mean temperatures
# mean_limits = find_common_limits(analysis_frame[axis_prefix.lower().replace(' ','_')+'_mean_outdoor'], 
#                                  analysis_frame[axis_prefix.lower().replace(' ','_')+'_mean_indoor'])

mean_limits = 50, 100

# # Custom colormap for NDVI
# cmap = sns.light_palette("green", as_cmap=True)


# Mean average temperatures - modified to color by mean_ndvi
ax1 = plt.subplot(121)
sns.scatterplot(
    data=analysis_frame[analysis_frame['Day/night'] == day_night], 
    x=axis_prefix.lower().replace(' ','_')+'_mean_outdoor', 
    y=axis_prefix.lower().replace(' ','_')+'_mean_indoor',
    hue='mean_ndvi',  # Color by mean NDVI value
    # palette=cmap,
    palette='PiYG',
    alpha=0.25,
    legend=False,  # Temporarily disable legend
    ax=ax1
)

# Add colorbar for NDVI
smc = plt.cm.ScalarMappable(cmap='PiYG', norm=plt.Normalize(vmin=analysis_frame['mean_ndvi'].min(), vmax=analysis_frame['mean_ndvi'].max()))
smc.set_array([])
cbar = plt.colorbar(smc, ax=ax1,fraction=0.046, pad=0.04)
cbar.set_label('Mean NDVI')

sns.regplot(data=analysis_frame[analysis_frame['Day/night'] == day_night], 
            x=axis_prefix.lower().replace(' ','_')+'_mean_outdoor', 
            y=axis_prefix.lower().replace(' ','_')+'_mean_indoor', 
            ax=ax1, 
            scatter = False,
            # scatter_kws={'alpha': 0.5},
            line_kws={'color': 'red'}, 
            ci=95)
ax1.set_xlim(mean_limits)
ax1.set_ylim(mean_limits)
ax1.set_aspect('equal')
ax1.plot(mean_limits, mean_limits, ls="--", c=".3")  # x=y line
ax1.set_title(f'{day_night}time Mean Humidity')
ax1.set_ylabel('Indoor Relative Humidity (%)', fontsize = 14)
ax1.set_xlabel('Outdoor Relative Humidity (%)', fontsize = 14)
annotate_fit(analysis_frame[axis_prefix.lower().replace(' ','_')+'_mean_outdoor'], analysis_frame[axis_prefix.lower().replace(' ','_')+'_mean_indoor'], ax1)

if day_night == 'Day':
    # max_limits = find_common_limits(analysis_frame[axis_prefix.lower().replace(' ','_')+'_max_outdoor'], 
    #                                 analysis_frame[axis_prefix.lower().replace(' ','_')+'_max_indoor'],
    #                                 upper_bound=100)  # Set upper bound to 50

    max_limits = 50, 100

    # Mean maximum temperatures
    ax2 = plt.subplot(122)
    sns.scatterplot(data=analysis_frame[analysis_frame['Day/night'] == day_night], 
                x=axis_prefix.lower().replace(' ','_')+'_max_outdoor', 
                y=axis_prefix.lower().replace(' ','_')+'_max_indoor',
                hue='mean_ndvi',  # Color by mean NDVI value
                # palette=cmap,
                palette='PiYG',
                alpha=0.25,
                legend=False,  # Temporarily disable legend
                ax=ax2)
    # Add colorbar for NDVI
    smc2 = plt.cm.ScalarMappable(cmap='PiYG', norm=plt.Normalize(vmin=analysis_frame['mean_ndvi'].min(), vmax=analysis_frame['mean_ndvi'].max()))
    smc2.set_array([])
    cbar2 = plt.colorbar(smc2, ax=ax2,fraction=0.046, pad=0.04)
    cbar2.set_label('Mean NDVI')
    sns.regplot(data=analysis_frame[analysis_frame['Day/night'] == day_night], 
                x=axis_prefix.lower().replace(' ','_')+'_max_outdoor', 
                y=axis_prefix.lower().replace(' ','_')+'_max_indoor', 
                ax=ax2, 
                # scatter_kws={'alpha': 0.5}, 
                scatter = False,
                line_kws={'color': 'red'}, 
                ci=95)
    ax2.set_xlim(max_limits)
    ax2.set_ylim(max_limits)
    ax2.set_aspect('equal')
    ax2.plot(max_limits, max_limits, ls="--", c=".3")  # x=y line
    ax2.set_title(f'{day_night}time Maximum Humidity')
    ax2.set_ylabel('Relative Humidity (%)', fontsize = 14)
    ax2.set_xlabel('Relative Humidity (%)', fontsize = 14)
    annotate_fit(analysis_frame[axis_prefix.lower().replace(' ','_')+'_max_outdoor'], analysis_frame[axis_prefix.lower().replace(' ','_')+'_max_indoor'], ax2)

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(bottom=-0.25)  # Provide space at the bottom for the legend

plt.savefig(f"../humidity_plots/{analysis_level}_{day_night}_correlation_indoor_outdoor_ndvi_humidity.png", bbox_inches='tight')
```
