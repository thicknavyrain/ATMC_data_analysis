```python
def calculate_frequency_and_std(dataframe, group_column):
    # Ensure the "Date/Time" column is in datetime format
    dataframe['Date/Time'] = pd.to_datetime(dataframe.index)
    
    # Calculate the difference in time between consecutive measurements within each group
    dataframe['Time_Diff'] = dataframe.groupby(group_column)['Date/Time'].diff().dt.total_seconds() / 60  # Convert to minutes
    
    # Define a function to get the smallest mode
    def smallest_mode(series):
        return series.mode().min()
    
    # Calculate the average frequency, standard deviation, count, and smallest mode for each group
    results = dataframe.groupby(group_column)['Time_Diff'].agg(['mean', 'std', 'count', ('mode', smallest_mode)]).reset_index()
    
    # Calculate the overall average of means and modal values
    overall_mean = results['mean'].mean()
    overall_std = results['std'].mean()
    overall_mode_avg = results['mode'].mean()
    
    # Print the results for each group
    for index, row in results.iterrows():
        print(f"Site ID: {row[group_column]}, Average Frequency: {row['mean']:.2f} minutes, Standard Deviation: {row['std']:.2f} minutes, Modal Frequency: {row['mode']:.2f} minutes, Data Points: {row['count']}")
    
    # Print the overall averages
    print(f"\nOverall Average Frequency (Mean): {overall_mean:.2f} minutes, Overall Average Standard Deviation: {overall_std:.2f} minutes, Overall Modal Average Frequency: {overall_mode_avg:.2f} minutes")

# calculate_frequency_and_std(outdoor_temp, 'site_id')
# calculate_frequency_and_std(indoor_temp, 'sensor_id_indoor_temp')
# 
calculate_frequency_and_std(outdoor_hum, 'site_id')
# calculate_frequency_and_std(indoor_hum, 'sensor_id_indoor_hum')
```

```python
import pandas as pd

# Assuming outdoor_temp, outdoor_hum, and outdoor_link_df are already loaded DataFrames

# Rename 'Value' column to specific names before setting the index
outdoor_temp.rename(columns={'Value': 'Value_temp'}, inplace=True)
outdoor_hum.rename(columns={'Value': 'Value_hum'}, inplace=True)

# Add 'site_id' to outdoor_temp and outdoor_hum by mapping from outdoor_link_df
# outdoor_temp['site_id'] = outdoor_temp['point_id_outdoor_temp'].map(outdoor_link_df.set_index('point_id_outdoor_temp')['site_id'])
# outdoor_hum['site_id'] = outdoor_hum['point_id_outdoor_hum'].map(outdoor_link_df.set_index('point_id_outdoor_hum')['site_id'])

# # Convert 'Date/Time' to a datetime index and ensure it is sorted for both DataFrames
# outdoor_temp.index = pd.to_datetime(outdoor_temp.index)
# outdoor_hum.index = pd.to_datetime(outdoor_hum.index)

outdoor_temp = outdoor_temp.sort_index()
outdoor_hum = outdoor_hum.sort_index()

# Define tolerance for merging (e.g., 10 minutes)
tolerance = pd.Timedelta(minutes=10)

# Merge using pd.merge_asof. Note that this requires both DataFrames to be sorted by the 'Date/Time' index.
# We'll perform this operation for each 'site_id' to ensure correctness.
result_frames = []
for site_id in outdoor_link_df['site_id'].unique():
    temp_df = outdoor_temp[outdoor_temp['site_id'] == site_id]
    hum_df = outdoor_hum[outdoor_hum['site_id'] == site_id]
    
    # Ensure data is sorted by 'Date/Time'
    temp_df = temp_df.sort_index()
    hum_df = hum_df.sort_index()
    
    # Reset index to bring 'Date/Time' back as a column for merging
    temp_df.reset_index(inplace=True)
    hum_df.reset_index(inplace=True)
    
    # Merge with pd.merge_asof
    merged_df = pd.merge_asof(temp_df, hum_df, on='Date/Time', by='site_id', tolerance=tolerance, direction='nearest')

    print(len(merged_df))
    result_frames.append(merged_df)

# Concatenate all the merged data for each site
final_merged_df = pd.concat(result_frames)

# Assuming you want to keep only rows where both temperature and humidity data are available
final_merged_df = final_merged_df.dropna(subset=['Value_temp', 'Value_hum'])

print(final_merged_df[['Date/Time', 'site_id', 'Value_temp', 'Value_hum']].head())
```
